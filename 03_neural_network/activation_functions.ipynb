# 신경망 (Neura Network)

## 활성화 함수 (Activation Function)

```python
### 계단함수(Step function)

import numpy as np
import matplotlib.pylab as plt

def step_function0(x):
    if x > 0:
        return 1
    else:
        return 0

def step_function1(x):
    y = x > 0
    return y.astype(np.int)

def step_function2(x):
    return np.array(x>0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function2(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
### 시그모이드 함수(Sigmoid Function) S자 모형 함수
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
### 시그모이드 함수와 계단 함수의 비교
import numpy as np
import matplotlib.pylab as plt

def step_function(x):
    return np.array(x > 0, dtype=int)
def sigmoid_function(x):
    return 1 / ( 1 + np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x)
y2 = sigmoid_function(x)
plt.plot(x, y1, 'g--', label='step_function')
plt.plot(x, y2, 'r-', label='sigmoid_function')
plt.show()
```

```python
### 렐루 함수 (Rectified Linear Unit)
import numpy as np
import matplotlib.pylab as plt

def relu(x):
    return np.maximum(0, x)
def step_function(x):
    return np.array(x > 0, dtype=int)
def sigmoid_function(x):
    return 1 / ( 1 + np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y1 = step_function(x)
y2 = sigmoid_function(x)
y3 = relu(x)
plt.plot(x, y2, 'r-', label='sigmoid_function')
plt.plot(x, y1, 'g--', label='step_function')
plt.plot(x, y3, 'b-.', label='relu')
plt.ylim(-0.1, 1.1)
plt.show()
```
